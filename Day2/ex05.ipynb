```python
from docx import Document
```

Local File

```python
file_path = "../Data/DWH_SEAMAP_Groundfish_Dataset_Documentation.docx"
```

- For Text

```python
# Document() constructor, which returns DocumentObject (This takes str, bytesIO, etc.)
document = Document(file_path) 

output = []
for paragraph in document.paragraphs:
    print(paragraph.text)
    output.append(paragraph.text)
```

- For Tables

```python
tables = []

# For tables
for table in document.tables:
    output = []
    for row in table.rows:
        row_content = [cell.text for cell in row.cells]  # Collect cell content in a list
        output.append(row_content)
    tables.append(output)
```

```python
# get request
response = requests.get(url)
# get content =>ByteIO
data = BytesIO(response.content)
# Document() to read data
document = Document(data)
document
```

```python
# For text
text_output = []
for paragraph in document.paragraphs:
    text_output.append(paragraph.text)
```

```python
# For tables
output_tables = []
for table in document.tables:
    output_table = []
    for row in table.rows:
        output_row = []
        for cell in row.cells:
            output_row.append(cell.text)
        output_table.append(output_row)
    output_tables.append(output_table)



# for table in document.tables:
#     output = []
#     for row in table.rows:
#         row_content = [cell.text for cell in row.cells]  # Collect cell content in a list
#         output.append(row_content)
#     tables += output
```
